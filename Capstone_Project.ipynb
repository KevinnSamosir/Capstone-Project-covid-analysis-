```python
df = pd.read_csv("country_wise_latest.csv")
```

```python
top_weekly_increase = df.sort_values(by="1 week % increase", ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(data=top_weekly_increase, x="1 week % increase", y="Country/Region", palette="flare")
plt.title("Top 10 Negara dengan Peningkatan Mingguan Tertinggi (%)")
plt.xlabel("Persentase Kenaikan Mingguan (%)")
plt.ylabel("Negara")
plt.tight_layout()
plt.show()
```

```python
region_impact = df.groupby("WHO Region")["Confirmed"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=region_impact.values, y=region_impact.index, palette="Reds_r")
plt.title("Total Kasus Terkonfirmasi per Wilayah WHO")
plt.xlabel("Jumlah Kasus Terkonfirmasi")
plt.ylabel("Wilayah WHO")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Deaths / 100 Cases", y="Recovered / 100 Cases", hue="WHO Region", palette="tab10", s=80)
plt.title("Korelasi antara Tingkat Kematian dan Kesembuhan")
plt.xlabel("Deaths per 100 Cases")
plt.ylabel("Recovered per 100 Cases")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
df["Prediksi minggu depan"] = df["Confirmed"] + df["1 week change"]
top_prediksi = df.sort_values(by="Prediksi minggu depan", ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(data=top_prediksi, x="Prediksi minggu depan", y="Country/Region", palette="Blues_d")
plt.title("Prediksi Total Kasus Minggu Depan (Top 10 Negara)")
plt.xlabel("Prediksi Jumlah Kasus")
plt.ylabel("Negara")
plt.tight_layout()
plt.show()
```
